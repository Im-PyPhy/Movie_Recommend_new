## Constructing the Data<br>
#### Steps Involved:<br>
1. **First a dataframe with the movies on the first page of the api is constructed and called 'df'.**<br>
2. Next we will loop through 2 to 501 pages in the api and concatenate each page by row with df to get the dataset for 10000 movies.<br>
3. Next we will add the cast and crew to each movies using the cast and crew api.<br>
4. We will convert the genre id to genres using the genre api.<br>
5. Adding the Keyword column by fetching the keywords of the movies from the keyword API.<br>
6. Dataset is completed and exported as csv.<br>

```python
## Importing required libraries:
import requests
from bs4 import BeautifulSoup
import pandas as pd
import time
import json
```

### Step 1.<br>
First a dataframe with the movies on the first page of the api is constructed and called 'df'.

```python
## First a dataframe with the movies on the first page of the api is constructed and called 'df'.
website = f'https://api.themoviedb.org/3/movie/top_rated?api_key=d580ed432d7d56f0eb97549c4fcb2273&language=en-US&page={1}'
response = requests.get(website).json()
df = pd.DataFrame(response.json()['results'])[['id','title','overview','original_language','release_date','vote_average','genre_ids']]
```

```python
df
```

### Step 2.<br>
Looping through 2 to 501 pages in the api and concatenate each page by row with df to get the dataset for 10000 movies.

```python
for i in range(2,501):
    website = f'https://api.themoviedb.org/3/movie/top_rated?api_key=d580ed432d7d56f0eb97549c4fcb2273&language=en-US&page={i}'
    response = requests.get(website).json()
    df2= pd.DataFrame(response['results'])[['id','title','overview','original_language','release_date','vote_average','genre_ids']]
    df = pd.concat([df,df2], axis=0)
```

```python
df
```

### Step 3. <br>
Adding the cast and crew to each movies using the cast and crew api.

```python
cast_df = pd.DataFrame(columns=('id','cast','crew'))
for i,id_ in enumerate(df['id']):
    if i%100==0:
        time.sleep(5)
    web = f"https://api.themoviedb.org/3/movie/{id_}/credits?api_key=d580ed432d7d56f0eb97549c4fcb2273&language=en-US"
    req = requests.get(web).json()
    alpha = {'id': req['id'],'cast':req['cast'],'crew':req['crew']} ## This is deltails of cast and crew of movie having movie id = id_
    cast_df = cast_df.append(alpha,ignore_index=True) # appending the cast and crew in a data frame cast_df 
```

```python
Movie_df_full=pd.merge(df,cast_df,on='id',how='inner') ## merging the cast_df with df on id column.
```

```python
Movie_df_full.drop('Unnamed: 0',axis=1,inplace=True)
```

```python
Movie_df_full.rename(columns={'id':'movie_id'},inplace=True)
```
